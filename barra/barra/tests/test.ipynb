```python
import polars as pl

lf = pl.LazyFrame(
    {
        "ticker_symbol": [1, 2, 3, 4, 5, 6],
        "b": [
            "2023-01-01",
            "2023-01-02",
            "2024-01-03",
            "2024-01-04",
            "2025-01-05",
            "2025-01-06",
        ],
        " c": [10, 20, 30, 40, 50, 60],
    },
    schema={"ticker_symbol": pl.Int64, "b": pl.String, " c": pl.Int64},
)
df = (
    lf.with_columns(pl.col("b").str.strptime(pl.Date, format="%Y-%m-%d"))
    .with_columns(
        pl.col("b").dt.year().rank("dense").alias("time_index"),
    )
    .collect()
)
df
```
