```python
import tushare as ts
from tqdm import tqdm
import os
```

```python
ts.set_token("4288b62ce66c47333d99e85b7c4987c63038341074a47a5392dab2b8")
pro = ts.pro_api()
pro.fund_basic(market="E").to_csv("/home/haris/project/etf/data/etf.csv", index=False)

df = pro.trade_cal(exchange="SSE", start_date="20230101", end_date="20251114")
trade_days = sorted(df[df.is_open==1]["cal_date"].tolist())
```

```python
# 读取etf净值（注意每分钟500次请求限制）
os.chdir(r"/home/haris/project/etf/data/daily_nav")
for date in tqdm(trade_days):
    if os.path.exists(f'{date}.csv'):
        continue
    df = pro.fund_nav(nav_date=date, market='E')
    df.to_csv(f'{date}.csv', index=False)
```

```python
os.chdir(r"/home/haris/project/etf/data/daily_share")
for date in tqdm(trade_days):
    if os.path.exists(f'{date}.csv'):
        continue
    df = pro.fund_share(trade_date=date)
    df = df[df['fund_type']=='ETF']
    df.to_csv(f'{date}.csv', index=False)
```

```python
os.chdir(r"/home/haris/project/etf/data/daily_factor")
for date in tqdm(trade_days):
    if os.path.exists(f'{date}.csv'):
        continue
    df = pro.fund_factor_pro(trade_date=date)
    df.to_csv(f'{date}.csv', index=False)
```
